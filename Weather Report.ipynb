```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#Make API Calls
api = '40e5eee38fb649a216b19ef663f9c0c9'
base_url = "https://api.openweathermap.org/data/2.5/weather?q="
temps = []
humids = []
winds = []
clouds = []
lats = []
names = []
ids = []
for city in cities:
    full_url = base_url + city + "&appid=" + api
    r = requests.get(full_url)
    data = r.json()
    try:
        humids.append(data['main']['humidity'])
        temps.append(data['main']['temp'])
        winds.append(data['wind']['speed'])
        clouds.append(data['clouds']['all'])
        lats.append(data['coord']['lat'])
        names.append(data['name'])
        ids.append(data['id'])
    except KeyError:
        print('none')
```

```python
for x in temps:
    c = (x-273.15)
    f = ((c*9)/5)+32
new_dict = {'city': names, 'city id': ids, "temp": temps, "humidity": humids, "wind speed": winds, "Cloud Coverage": clouds, 'lat': lats}
df = pd.DataFrame(new_dict)
df
```
